```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/media/WoW.xlsx'
df = pd.read_excel(file_path)

print(df.head())

df['SEMANA'] = pd.to_datetime(df['SEMANA'])

df['receita'] = df['receita'] / 100

grouped_df = df.groupby(['BU', 'SEMANA']).sum().reset_index()

unique_bus = grouped_df['BU'].unique()

colors = ['#e8175d', '#474747', '#0b2e59']

for bu in unique_bus:
    bu_data = grouped_df[grouped_df['BU'] == bu]

    plt.figure(figsize=(10, 8))

    # Sess√µes
    plt.subplot(3, 1, 1)
    sessoes = bu_data['sess√µes']
    pct_sessoes = sessoes.pct_change() * 100
    plt.plot(bu_data['SEMANA'], sessoes, marker='o', label='sess√µes', color=colors[0])
    for i, txt in enumerate(sessoes):
        plt.annotate(f"{txt}\n({pct_sessoes.iloc[i]:.2f}%)", (bu_data['SEMANA'].iloc[i], sessoes.iloc[i]), textcoords="offset points", xytext=(0,10), ha='center')
    plt.title(f'{bu} - Sess√µes por Semana')
    plt.xlabel('Semana')
    plt.ylabel('Sess√µes')
    plt.legend()
    plt.grid(False)

    # Pedidos
    plt.subplot(3, 1, 2)
    pedidos = bu_data['pedidos']
    pct_pedidos = pedidos.pct_change() * 100
    plt.plot(bu_data['SEMANA'], pedidos, marker='o', color=colors[1], label='pedidos')
    for i, txt in enumerate(pedidos):
        plt.annotate(f"{txt}\n({pct_pedidos.iloc[i]:.2f}%)", (bu_data['SEMANA'].iloc[i], pedidos.iloc[i]), textcoords="offset points", xytext=(0,10), ha='center')
    plt.title(f'{bu} - Pedidos por Semana')
    plt.xlabel('Semana')
    plt.ylabel('Pedidos')
    plt.legend()
    plt.grid(False)

    # Receita
    plt.subplot(3, 1, 3)
    receita = bu_data['receita']
    pct_receita = receita.pct_change() * 100
    plt.plot(bu_data['SEMANA'], receita, marker='o', color=colors[2], label='receita')
    for i, txt in enumerate(receita):
        plt.annotate(f"R${receita.iloc[i]:,.2f}\n({pct_receita.iloc[i]:.2f}%)", (bu_data['SEMANA'].iloc[i], receita.iloc[i]), textcoords="offset points", xytext=(0,10), ha='center')
    plt.title(f'{bu} - Receita por Semana')
    plt.xlabel('Semana')
    plt.ylabel('Receita (R$)')
    plt.legend()
    plt.grid(False)

    plt.tight_layout()
    plt.show()

    # Descri√ß√£o detalhada de cada gr√°fico
    print(f"Gr√°fico para {bu}:")
    print(f"- Sess√µes: Mostra a varia√ß√£o semanal do n√∫mero de sess√µes.")
    print(f"- Pedidos: Indica a quantidade de pedidos realizados a cada semana.")
    print(f"- Receita: Representa a receita gerada semanalmente em reais.")
    print("\n")

# Breve resumo do que os n√∫meros representam
print("Resumo:")
print("- Sess√µes: Refere-se ao n√∫mero total de visitas ao site durante a semana especificada.")
print("- Pedidos: Representa a quantidade total de pedidos feitos no site durante a semana.")
print("- Receita: Indica o total de receita gerada pelas vendas no site durante a semana, em reais.")
```
